```
from gensim.models import KeyedVectors
import gensim.models.word2vec

model = gensim.models.KeyedVectors.load_word2vec_format('news_mystem_skipgram_1000_20_2015.bin.gz', binary=True, encoding = 'utf8')



  
```

```
words = 'усомниться, страдать, сомневаться, сожалеть, раскритиковать, противоречить, протестовать, препятствовать, посетовать, пожаловаться, подозревать, отказывать, отказать, отказывать, отвергать, отвергнуть, осуждать, осудить, обеспокоить, критиковать, запрещать, запретить, жаловаться, грозить, бояться, беспокоить'.split(', ')
for word in words:
  print('{}\n'.format(word))
  try:
      a = model.most_similar('{}_V'.format(word))
      for b in a:
        print(b)
      print('\n')
  except:
    print('слова нет')
    print('\n')
    continue
```

```
words = 'вдохновить, верить, гордиться, доверять, заинтересоваться, интересоваться, любить, надеяться, нравиться, одобрить, одобрять, поверить, понравиться, похвалить, приветствовать, уважать'.split(', ')
for word in words:
  print('{}\n'.format(word))
  try:
      a = model.most_similar('{}_V'.format(word))
      for b in a:
        print(b)
      print('\n')
  except:
    print('слова нет')
    print('\n')
    continue
```
