```
from gensim.models import KeyedVectors
import gensim.models.word2vec

model = gensim.models.KeyedVectors.load_word2vec_format('model.bin', binary=True)
```

```
words = "усомниться страдать сомневаться сожалеть раскритиковать противоречить протестовать препятствовать посетовать пожаловаться подозревать отказывать отказать отказывать отвергать отвергнуть осуждать осудить обеспокоить критиковать запрещать запретить жаловаться грозить бояться беспокоить вдохновить верить гордиться доверять заинтересоваться интересоваться любить надеяться нравиться одобрить одобрять поверить понравиться похвалить приветствовать уважать".split()

for word in words:
    # есть ли слово в модели? Может быть, и нет
    word = word + "_VERB"
    if word in model:
        print(word)
        # выдаем 10 ближайших соседей слова:
        for i in model.most_similar(positive=[word], topn=10):
            # слово + коэффициент косинусной близости
            print(i[0], i[1])
        print('\n')
    else:
        # Увы!
        print(word + ' is not present in the model')
```
