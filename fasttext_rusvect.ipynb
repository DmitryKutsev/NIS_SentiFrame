```
import gensim
import numpy as np
import pandas as pd

import warnings
#warnings.filterwarnings('ignore')
```

```
model = gensim.models.KeyedVectors.load("araneum_none_fasttextcbow_300_5_2018.model")
```

```
WORDS = """усомниться, страдать, сомневаться, сожалеть, раскритиковать, противоречить, протестовать, препятствовать, посетовать, пожаловаться, подозревать, отказывать,отказать, отказывать, отвергать, отвергнуть, осуждать, осудить, обеспокоить, критиковать, запрещать, запретить, жаловаться, грозить, бояться, беспокоить, вдохновить, верить, гордиться, доверять, заинтересоваться, интересоваться, любить, надеяться, нравиться, одобрить, одобрять, поверить, понравиться, похвалить, приветствовать, уважать"""
```

```
words_list = [word.strip(' ') for word in WORDS.split(',')]
```

```
model.most_similar('усомниться')
```

```
def most_similar_words(word, model):
  try:
    vec = model.most_similar(word)
    return {word: vec}
  except:
    return {word: 0}

def eval_words(words_list, model):
  result = []
  for word in words_list:
    result.append(most_similar_words(word, model))
  return result
```

```
my_words = eval_words(words_list, model)
```
